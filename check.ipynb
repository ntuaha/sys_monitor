## 信用卡冒用偵測

```python
a = [(t,n) for t,n in df.iloc[10,25:].reset_index().values if n is not np.nan]
```

```python
def parser(name):
    rows = name.split('\n')
    cols = {}
    for row in rows:
        if row=='' or '>1秒次數' in row:
            continue
        if '~' not in row:
            continue
        try:
            k,v = row.split(':')
        except ValueError:
            continue
        cols[k] = int(v.replace(' ','').replace('\xa0','').replace('s','').replace(',','').replace('.',''))
    return cols

rows = [0]*len(a)
for i in range(len(a)):
    row = parser(a[i][1])
    row['t'] = a[i][0]
    rows[i] = row
```

```python
import scipy.stats
```

```python
weight = np.linspace(0.05,0.95,10)
```

```python
keys = list(rows[6].keys())[:-1]
```

```python
# 檢定分佈是否偏離
```

```python
def gen_data(rows,keys,num_a,num_b):
    
    row_b = [rows[num_b][k] for k in keys]
    
    return row_a,row_b
row_a,row_b = gen_data(rows,keys,13,14)
```

```python
dd1 = []
for i in range(2,len(row_a)):
    dd1 += row_a[i]*[weight[i]]

dd2 = []
for i in range(2,len(row_a)):
    dd2 += row_b[i]*[weight[i]]
```

```python
from scipy.stats import mannwhitneyu
```

```python
u_statistic, p = mannwhitneyu(dd1, dd2,use_continuity = True, alternative = 'two-sided')
print('U statistic:', u_statistic)
print('p-value:', p)
if p >= 0.05:
    print('H0 is accepted')
else:
    print('H0 is rejected')
```

```python
def gen_mean(rows,keys,weight,num):
    try:
        row = [rows[num][k] for k in keys]
        mean = np.sum(row * weight) / np.sum(row)
    except:
        return None
    return mean

```
